```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
import mysql.connector
from datetime import datetime, timedelta
from googleapiclient.discovery import build
import requests
import pytz

# YouTube API 설정
API_KEY = "AIzaSyC2TpuVyKa5AM8TrnJ4EVxDiJ4zU5H0zrk"
youtube = build("youtube", "v3", developerKey=API_KEY)

# 카테고리 ID와 이름 매핑을 위한 API 호출
def fetch_category_mapping(api_key):
    url = f"https://www.googleapis.com/youtube/v3/videoCategories?part=snippet&regionCode=US&key={api_key}"
    response = requests.get(url)
    data = response.json()
    category_mapping = {item['id']: item['snippet']['title'] for item in data['items']}
    return category_mapping

category_mapping = fetch_category_mapping(API_KEY)

# 요일을 한국어로 정의
days_of_week_kr = {
    '월요일': 0,
    '화요일': 1,
    '수요일': 2,
    '목요일': 3,
    '금요일': 4,
    '토요일': 5,
    '일요일': 6
}

# 오늘 날짜 계산 (KST)
kst = pytz.timezone('Asia/Seoul')
today = datetime.now(kst)

def get_date_from_day(day):
    today_weekday = today.weekday()
    target_weekday = days_of_week_kr[day]
    # 이번 주의 해당 요일 날짜 계산
    start_of_week = today - timedelta(days=today_weekday)
    if target_weekday <= today_weekday:
        target_date = start_of_week + timedelta(days=target_weekday)
    else:
        target_date = start_of_week + timedelta(days=(target_weekday - 7))
    return f'{target_date.month}월 {target_date.day}일'

def get_video_category(video_id):
    try:
        request = youtube.videos().list(
            part="snippet",
            id=video_id
        )
        response = request.execute()
        if "items" in response and len(response["items"]) > 0:
            category_id = response["items"][0]["snippet"]["categoryId"]
            category_name = category_mapping.get(category_id, "Unknown")
            return category_id, category_name
    except Exception as e:
        print(f"Error fetching category for video {video_id}: {e}")
    return None, None

def fetch_youtube_history():
    options = Options()
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option('useAutomationExtension', False)


    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)
    driver.get("https://www.youtube.com/feed/history")
    time.sleep(60)  # 로그인 시간

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)

    data = []

    sections = driver.find_elements(By.XPATH, '//ytd-item-section-renderer')
    for i, section in enumerate(sections, start=1):
        try:
            videos = section.find_elements(By.TAG_NAME, 'ytd-video-renderer')
            for j, video in enumerate(videos, start=1):
                try:
                    title_element = video.find_element(By.ID, 'video-title')
                    channel_element = video.find_element(By.XPATH, f'.//div[1]/div/div[1]/ytd-video-meta-block/div[1]/div[1]/ytd-channel-name/div/div/yt-formatted-string/a')
                    view_count_element = video.find_element(By.XPATH, './/span[contains(@class, "style-scope ytd-video-meta-block")]')
                    date_element = section.find_element(By.XPATH, f'.//ytd-item-section-header-renderer/div[1]/div')
                    watch_date = date_element.text.strip()

                    # Watch Date가 요일로만 표시된 경우
                    if watch_date in days_of_week_kr:
                        watch_date = get_date_from_day(watch_date)
                    elif watch_date == "오늘":
                        watch_date = f'{today.month}월 {today.day}일'

                    video_url = title_element.get_attribute('href')
                    if video_url is not None and "watch?v=" in video_url:
                        video_id = video_url.split('watch?v=')[-1].split('&')[0]
                        category_id, category_name = get_video_category(video_id)
                    else:
                        video_id = "Unknown"
                        category_id, category_name = None, None

                    if category_id is None or category_name is None:
                        print(f"Failed to fetch category for video ID: {video_id}")

                    data.append({
                        'Video ID': video_id,
                        'Channel Name': channel_element.text.strip(),
                        'Video Title': title_element.text.strip(),
                        'Views': view_count_element.text.strip(),
                        'Watch Date': watch_date,
                        'Category ID': category_id if category_id is not None else "Unknown",
                        'Category Title': category_name if category_name is not None else "Unknown"
                    })
                except Exception as e:
                    print(f"Error extracting video info: {e}")
                    continue
        except Exception as e:
            print(f"Error extracting section info: {e}")
            continue

    driver.quit()
    return data

def save_to_database(data, db_config):
    db = mysql.connector.connect(**db_config)
    cursor = db.cursor()

    cursor.execute("""
    CREATE TABLE IF NOT EXISTS videos (
        id INT AUTO_INCREMENT PRIMARY KEY,
        video_id VARCHAR(255),
        channel_name VARCHAR(255),
        video_title VARCHAR(255),
        views VARCHAR(100),
        watch_date VARCHAR(100),
        category_id VARCHAR(255),
        category_title VARCHAR(255)
    )
    """)

    for item in data:
        sql = "INSERT INTO videos (video_id, channel_name, video_title, views, watch_date, category_id, category_title) VALUES (%s, %s, %s, %s, %s, %s, %s)"
        val = (item['Video ID'], item['Channel Name'], item['Video Title'], item['Views'], item['Watch Date'], item['Category ID'], item['Category Title'])
        cursor.execute(sql, val)

    db.commit()
    print(cursor.rowcount, "records inserted.")
    cursor.close()
    db.close()

# 주 실행 함수
def main():
    db_config = {
        'host': 'localhost',
        'user': 'root',
        'password': 'pw1234',
        'database': 'youtube_history'
    }
    data = fetch_youtube_history()
    save_to_database(data, db_config)

if __name__ == "__main__":
    main()
```
